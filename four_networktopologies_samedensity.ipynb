```python
import time, enum, math
from collections import defaultdict
from typing import Tuple

import numpy as np
import networkx as nx
import matplotlib as plt
```

```python

graph = nx.connected_caveman_graph(10, 10)
nx.draw_kamada_kawai(graph)
"total number is equal to k*l"
"https://networkx.org/documentation/stable/reference/generated/networkx.generators.community.connected_caveman_graph.html"
clust=nx.average_clustering(graph, nodes=None, weight=None, count_zeros=True)
print(clust)
dens=nx.density(graph)
print(dens)
```

```python
graph=nx.windmill_graph(10,11)
nx.draw_kamada_kawai(graph)
"total number is not equal to k*(l-1)+1"
"https://networkx.org/documentation/stable/reference/generated/networkx.generators.community.windmill_graph.html"
clust=nx.average_clustering(graph, nodes=None, weight=None, count_zeros=True)
print(clust)
dens=nx.density(graph)
print(dens)
```

```python
graph=nx.watts_strogatz_graph(100, 10, 0.5)
nx.draw_circular(graph)
"total number is equal n. second number refers to number of connections of each before stochaic.third number refers to the probability of stablishing a new extra connection stochaicly. previous connections are deleted, so that the number of connections is the same"
clust=nx.average_clustering(graph, nodes=None, weight=None, count_zeros=True)
print(clust)
dens=nx.density(graph)
print(dens)
```

```python
graph=nx.dual_barabasi_albert_graph(100,20,80,0.85)
nx.draw_kamada_kawai(graph)
"total number is n, first number. second number=number of edges at probability p. 3rd number: number of edges prob (p-1). 4th num p."
"https://networkx.org/documentation/stable/reference/generated/networkx.generators.random_graphs.dual_barabasi_albert_graph.html#networkx.generators.random_graphs.dual_barabasi_albert_graph"
clust=nx.average_clustering(graph, nodes=None, weight=None, count_zeros=True)
print(clust)
dens=nx.density(graph)
print(dens)
```

# The two first are "fixed" networks, the two latest are random. but all of them show the same density, even though very different clustering coefficients.
