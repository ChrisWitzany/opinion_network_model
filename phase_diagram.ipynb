```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage.filters import gaussian_filter

x = np.linspace(0, 1, 51)
y = np.linspace(0, 100, 51)
r = np.random.RandomState(42)

#z will eventually be the matrix containing at entry z(i,j) the value between 0 and 1 where 0 is no believers and 1 is 100% believers
z = gaussian_filter(r.random_sample([50, 50]), sigma=7, mode='wrap')
z -= np.min(z)
z /= np.max(z)

def plot_phase_diagram(x,y,z):
    fig, ax = plt.subplots()
    cmap = ax.pcolormesh(x, y, z)
    fig.colorbar(cmap)
    plt.xlabel("Conspiracy spreading probability")
    plt.ylabel("Number of initial believers")
    plt.show(fig)

plot_phase_diagram(x,y,z)
```
